```python
#map to grid function returns grid no of which the lat and lon belong
#grid to cord does the opposite
#path function see the origin and then sees which grid prob is the highest and then append it and assign it as origin for comparsion again
import operator
def map_to_grid(tup,arr):
    for j in range(len(arr)):
        if arr[j][0]+0.25>tup[0] and arr[j][1]+0.25>tup[1]:
            return j
def path(origin,probab,ts):
    arr=[]
    arr.append(origin)
    for i in range(ts):
        keyMax = max(probab[origin].items(), key = operator.itemgetter(1))[0] 
        arr.append(keyMax)
        origin=keyMax
    return arr
def grid_to_cord(val,arr):
    return arr[val]
```

```python
import numpy as np
 
class MarkovChain(object):
    def __init__(self, transition_prob):
        self.transition_prob = transition_prob
        self.states = list(transition_prob.keys())
 
    def next_state(self, current_state):
        p=[self.transition_prob[current_state][next_state] 
               for next_state in self.states]
        if sum(p)==0:
            return
        return np.random.choice(
            self.states, 
            p=p
        )
 
    def generate_states(self, current_state, no=10):
        future_states = []
        for i in range(no):
            next_state = self.next_state(current_state)
            if next_state==None:
                continue
            future_states.append(next_state)
            current_state = next_state
        return future_states
```
